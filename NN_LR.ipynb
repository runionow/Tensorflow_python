# Neural Network : Linear Regression ( Large Dataset + Noise )

```python
import tensorflow as tf
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
x_data = np.linspace(0.0,10.0,100000) + np.random.randn(100000)
```

y = mx + b

```python
y_true = (0.5*x_data)+ 5 + np.random.randn(100000)
```

```python
x_df = pd.DataFrame(data = x_data,columns =['XData'])
print(x_df.head(5))
```

```python
y_df = pd.DataFrame(data=y_true,columns=['YLabel'])
print(y_df.head(5))
```

```python
data = pd.concat([x_df,y_df],axis=1)
data.head(5)
```

```python
data.sample(n=250).plot(kind= 'scatter' ,x = "XData",y = "YLabel")
```

```python
batch_size = 8
```

```python
np.random.randn(2)
```

```python
m = tf.Variable(0.39)
b = tf.Variable(0.23)
```

```python
xph = tf.placeholder(tf.float32,[batch_size])
```

```python
yph =  tf.placeholder(tf.float32,[batch_size])
```

```python
y = m * xph + b
```

```python
# Loss Function.
error = tf.reduce_sum(tf.square(yph-y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.001)
```

```python
train = optimizer.minimize(error)
```

```python
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    batches = 1000
    for i in range(batches):
        rand_ind = np.random.randint(low = 0 , high = len(x_data),size = batch_size)
        print(x_data)
        feed = {xph:x_data[rand_ind],yph:y_true[rand_ind]}
        sess.run(train,feed_dict=feed)
    model_m,model_b =sess.run([m,b])
    # print(sess.run(error))
```

```python
print(model_m)
print(model_b)
```

```python
y_hat = x_data*model_m + model_b
```

```python
data.sample(n=250).plot(kind= 'scatter' ,x = "XData",y = "YLabel")
plt.plot(x_data,y_hat,'r')
```
